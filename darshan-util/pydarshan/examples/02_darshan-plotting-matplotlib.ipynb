## Plotting DarshanReports using `matplotlib`(Experimental)

This notebook demonstrates how to visualize Darshan reports using matplotlib.
The included plots are still experimental for 

```python
import darshan
darshan.enable_experimental()
```

```python
# merging records
from darshan.experimental.plots.matplotlib import plot_access_histogram
from darshan.experimental.plots.matplotlib import plot_opcounts

# prepare some reports
r1 = darshan.DarshanReport("example-logs/example.darshan", read_all=True)
r2 = darshan.DarshanReport("example-logs/ior_hdf5_example.darshan", read_all=True)
rx = r1 + r2
```

### Access Sizes (Binned into different categories 0-100 bytes, 101-1K, ..., 1G+)

```python
for r in [r1, r2, rx]:
    plt = plot_access_histogram(r, mod='POSIX')
    plt.show()
```

### Operation Counts (aggregated by type: read, write, open, ...)

```python
for r in [r1, r2, rx]:
    plt = plot_opcounts(r)
    plt.show()
```
